```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("GINF_DATA.csv",encoding='latin1')
df.head(5)
```

```python
plt.figure(figsize=(6,5))
ax = sns.countplot(data =df,x="sex")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.barplot(x='internet', y='Note_Final', data=df, ci=None, palette='viridis')

plt.xlabel('AccÃ¨s Ã  Internet')
plt.ylabel('Moyenne des notes')
plt.title('Impact de l\'accÃ¨s Ã  Internet sur les notes')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```
