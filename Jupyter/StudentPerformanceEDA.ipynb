# Model Building

* linear Model

```python
from sklearn.model_selection import train_test_split
```

```python
df.info()
```

```python
dr = df.drop(['Last Name',"First Name","Numero appoge","City"],axis=1)
x = dr.drop(["NOFINAL"],axis=1)
y = dr["NOFINAL"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
pred = model.predict(x_test)
output = pd.DataFrame({'actual_charges':y_test,'predicted_charges':pred})
output.head(30)
```

* Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor
from time import perf_counter

t1 = perf_counter()
model = RandomForestRegressor()

model.fit(x_train,y_train)

pred = model.predict(x_test)
output = pd.DataFrame({'actual_charges':y_test,'predicted_charges':pred})
output.head(30)
print(f"time taken: {perf_counter()-t1}")


#plt.scatter(x_test, y_test, color='black', label='Actual')
# plt.scatter(x_test, pred, color='red', label='Predicted')
plt.plot(y_test , pred , color="b")
plt.title('Random Forest')
plt.xlabel('y pred')
plt.ylabel('y actual')
plt.legend()
plt.show()
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.svm import SVR
from time import perf_counter

t1 = perf_counter()
model = SVR(kernel='linear', C=1.0, epsilon=0.2)

model.fit(x_train,y_train)


pred = model.predict(x_test)
output = pd.DataFrame({'actual_charges':y_test,'predicted_charges':pred})
output.head(30)
print(f"time taken: {perf_counter()-t1}")
```
