```python
plt.figure(figsize=(6,5))
ax = sns.countplot(data =df,x="sex")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
filiere = df['Filier'].value_counts()
filiere
```

```python
plt.figure(figsize=(10,7))
plt.bar(filiere.index,filiere.values,linewidth=1,edgecolor='black')
plt.title('Nombre des etudiants de chaque Filier',fontname = 'monospace', weight='bold',size=15)
plt.xlabel('Filiere',weight='semibold', fontname = 'monospace',size=10)
plt.ylabel('Nombre',weight='semibold', fontname = 'monospace',size=10)
plt.xticks(rotation=80)
plt.show()
```
